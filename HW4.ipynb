```python
from pymystem3 import Mystem

with open('Harry_Potter_and_the_Sorcerers_Stone.txt','r', encoding='utf-8') as f:
    text = f.read()
```

```python
from string import punctuation
punctuation += "’‘"
n = [w.strip(punctuation) for w in text.split()]
m = [w for w in n if w != "-"]

with open('Harry_new.txt', 'w') as f:
    f.write(' '.join(m))
```

```python
s = ' '.join(m)
with open('Harry_new.txt', 'w') as f:
    f.write(s.lower())
```

```python
text = s.lower()
from nltk import WordNetLemmatizer
wnl = WordNetLemmatizer()
q = wnl.lemmatize(text)
with open ('Harry_lemmas.txt', 'w', encoding = 'utf-8') as f:
    f.write(q)
```

```python
import collections
text = q.split()
c = collections.Counter()
for w in text:
    c[w] += 1
```

```python
import matplotlib.pyplot as plt
def tsipf(X, Y):
    return plt.plot(X, Y)
   
X = range(len(c))
Y = sorted([c/len(text) for c in c.values()], reverse=True)
    
```

```python
plt.plot(X, Y)
```
